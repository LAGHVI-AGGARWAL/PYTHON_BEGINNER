```python
# RANGE FUNCTTION
# ALLOWS USERS TO GENERATE A SERIES OF NUMBERS
# IT IS ITERABLE AS WELL

# RANGE(N)
# HERE N IS AN INTEGER
# THE RANGE FUNCTION TAKES THREE PARAMETERS - START, END , JUMP
# WHEN ONLY ONE PARAMETER IS PASSED IN THE RANGE FUNCTION THEN IT DENOTES THE END .
# THE START WOULD BY DEFAULT BE 0 AND JUMP WOULD BE 1
# THE START IS INCLUDED AND END IS EXCLUDED SO IT WILL PRINT INTEGER FROM
# START TO (END - 1)
```

```python
# break,continue,pass
# these statements are used to alter the flow of the program
# break : break the flow of the program once the condition is hit
# continue : it skips the particular iteration
# pass : to avoid syntax error
```

```python
for i in range(1,10):
  pass
  # it will not give any syntax error otherwise it would give as we have not written anything inside the indented block
```

```python
for i in range(1,10):
  if (i == 5):
    continue
  print(i)
```

```python
for i in range(1,10):
  if (i == 5):
    break
  print(i)
```
