```python
# print all the vowels of a given string
me = input()
me = me.lower()
for chr in (me[:]):
  if(chr == 'a' or chr == 'e' or chr == 'i' or chr == 'o' or chr == 'u'):
    print(chr, end = ' ')
```

```python
# check if the string is palindrome or not
# a string is said to be palindrome if the reverse of the string is equal
# to the original string
me = input()
me1 = me[-1:-(len(me)) - 1:-1]
if (me == me1):
  print(f"yes {me} is palindrome")
else:
  print(f"{me} is not palindrome")
```

```python
# count frequency of each character
me = input()
# here repeated data is coming
# for chr in me:
#   print(chr,me.count(chr))

# another method
seen = ''
for chr in me:
  if chr not in seen:
    print(chr,me.count(chr))
    seen += chr
```

```python
# printing string in uppercase without using upper method
# Every lowercase letter and its uppercase version are exactly 32 units apart in
#  the Unicode table. If you know the code for 'a', you can find 'A' by simply subtracting 32.
me = input()
new = ''
for char in me:
  if char.islower():
    char1 = chr(ord(char) - 32)
    new += char1
  else:
    new += char
print(new)
```
