```python
# How it works internally
# Internally, a Python list doesn't "hold" the actual data.
# Instead, it holds a sequence of references (memory addresses).
# Each reference points to an object (like the integer 123 or the string "ABC")
#  stored somewhere else in your computer's RAM. Because all these references
# are the same size, Python doesn't care if the objects they point to are different types.
```

```python
# list comprehension
# it is a way to create a new list by performing an operation on an existing one—all in a single line of code.
# new_list = [expression for item in iterable if condition]
# Expression: What you want to do to the item (The SELECT part).
# Item: The variable representing each element (The FROM part).
# Iterable: The original list/range you are looping through.
# Condition: (Optional) A filter to skip certain items (The WHERE part).
```

```python
# to calculate the square of an even number
# old way
numbers = [1, 2, 3, 4, 5]
numbers+= [6,7]
# squares = []
# for n in numbers:
#     if n % 2 == 0:
#         squares.append(n**2)
# print(squares)
print(numbers)
# using list comprehension
squares = [n**2 for n in numbers if n % 2 == 0]
print(squares)
```

```python
# When you use else, the logic moves to the front of the comprehension:
# [value_if_true if condition else value_if_false for item in iterable]
```

```python
numbers = range(1, 6)
labels = ["Even" if i % 2 == 0 else "Odd" for i in numbers]

print(labels)
```

```python
l = [1,2,3]
sum = 0
for i in l:
  sum+=i
print(sum)
```
